# **Overview**

**Geomorphic Zonation** refers to the division of a landscape into distinct zones based on morphological, structural, and process-related characteristics.

These zones are identified by variations in topography, landforms, erosion patterns, sediment deposition, and tectonic influences.

Remote sensing plays a crucial role in delineating these zones by providing synoptic, multi-spectral, and multi-temporal data for large and inaccessible areas.

# **Remote Sensing Techniques for Geomorphic Zonation**

Remote sensing enables the systematic classification of geomorphic zones using various data sources and analytical methods:

**A. Data Sources**

1. **Optical Remote Sensing (Multispectral & Hyperspectral)**

* Sensors: Landsat, Sentinel-2, MODIS, ASTER, WorldView

* Applications: Lithological discrimination, vegetation cover analysis, and landform mapping.

* Key spectral bands: SWIR (for rock composition), NIR (for vegetation and moisture), and TIR (for thermal properties).

2. **Radar Remote Sensing (SAR – Synthetic Aperture Radar)**

* Sensors: Sentinel-1, ALOS PALSAR, RADARSAT

* Applications: Surface roughness mapping, drainage patterns, and tectonic lineament detection (penetrates clouds and vegetation).

3. **LiDAR (Light Detection and Ranging)**

* Airborne & spaceborne LiDAR (e.g., ICESat-2)

* Applications: High-resolution DEM generation, micro-topographic feature extraction (e.g., river terraces, fault scarps).

4. **Digital Elevation Models (DEMs)**

* Sources: SRTM, ASTER GDEM, TanDEM-X, ALOS AW3D

* Applications: Slope, aspect, curvature analysis, and watershed delineation.

**B. Analytical Methods**

1. **Morphometric Analysis**

* Derived from DEMs: Slope, aspect, relief, hypsometric integral, and drainage density.

* Used to classify zones into mountains, plateaus, plains, or valleys.


2. **Spectral Indices & Image Classification**

* NDVI (vegetation cover), NDWI (water bodies), and mineralogical indices (e.g., Clay Mineral Ratio).

* Supervised (Maximum Likelihood, SVM) and unsupervised (ISODATA, K-means) classification for landform mapping.


3. **Texture & Pattern Analysis**

* GLCM (Gray-Level Co-occurrence Matrix) for surface roughness.

* Used to distinguish alluvial fans, dunes, or lava flows.


4. **Geomorphometric Modeling**

* Terrain Ruggedness Index (TRI), Topographic Position Index (TPI).

* Identifies ridges, valleys, and transitional zones.


5. **Machine Learning & AI**

* Deep learning (CNNs) for automated landform detection.

* Random Forest for integrating multi-source data (DEM + spectral + SAR).

# **Applications of Geomorphic Zonation**


1. **Tectonic Geomorphology**

Identifies fault zones, uplifted terraces, and folded regions using InSAR and DEM-derived lineaments.

2. **Fluvial & Coastal Geomorphology**

River channel migration, deltaic zonation, and coastal erosion monitoring using time-series satellite data.

3. **Glacial & Periglacial Zones**

Delineation of moraines, cirques, and permafrost features using thermal and LiDAR data.

4. **Arid & Desert Landscapes**

Sand dune classification, playa lakes, and yardangs using SAR backscatter and spectral indices.

5. **Volcanic Terrain Mapping**

Lava flows, calderas, and pyroclastic deposits mapped using thermal and high-resolution imagery.

# **Case Study**

The Kenyan coast, particularly Mombasa County's shoreline along the Indian Ocean, exhibits diverse geomorphic features shaped by marine, fluvial, and anthropogenic processes.

This study employs remote sensing techniques to delineate geomorphic zones, including beaches, coral reefs, estuaries, and coastal plains, while assessing erosion, accretion, and human impacts.

**Location**: Southern Kenya, bordering the Indian Ocean **(~4.05°S, 39.75°E)**.

**Key Features**:

i. Beaches & Dunes: Nyali, Bamburi, Diani (sandy, influenced by monsoons).

ii. Coral Reefs: Fringing reefs (e.g., Mombasa Marine Park).

iii. Estuaries & Creeks: Tudor Creek, Mtwapa Creek (mangrove ecosystems).

iv. Coastal Plains: Sandy and alluvial deposits, some lateritic uplands.

## Visualization

The Allen Coral Atlas dataset maps the geomorphic zonation and benthic habitat for the world's shallow coral reefs at 5 m pixel resolution. Also included is a global reef extent product that maps additional reef areas unable to be explicitly included in the geomorphic and benthic mapping.

The underlying satellite image data are temporal composites of PlanetScope satellite imagery spanning 2018-2020. The habitat maps are created via a machine learning approach with contextual editing, using a range of imagery, bathymetry and derived products as input data, trained via a globally consistent reference data set.

A full description of the methods and approaches can be found in the methods section of the Allen Coral Atlas website.

```python
dataset = ee.Image('ACA/reef_habitat/v2_0')
```

```python
geomorphicZonation = dataset.select('geomorphic').selfMask()
```

```python
Map.addLayer(geomorphicZonation, {}, 'Geomorphic Zonation')

ee_class_table = """
Value	Color	Description
0	#000000	Unmapped
11	#77d0fc	Shallow Lagoon
12	#2ca2f9	Deep Lagoon
13	#c5a7cb	Inner Reef Flat
14	#92739d	Outer Reef Flat
15	#614272	Reef Crest
16	#fbdefb	Terrestrial Reef Flat
21	#10bda6	Sheltered Reef Slope
22	#288471	Reef Slope
23	#cd6812	Plateau
24	#befbff	Back Reef Slope
25	#ffba15	Patch Reef

"""
legend_dict = geemap.legend_from_ee(ee_class_table)
Map.add_legend(title="Geomorphic Zoning", legend_dict=legend_dict)

Map
```
